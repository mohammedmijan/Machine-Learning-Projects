```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
train_data.info()
```

```python
object_column = []
float_column = []
int_column = []
for column in train_data.columns:
    if train_data[column].dtype == object:
        object_column.append(column)
    if train_data[column].dtype == "int64":
        int_column.append(column)
    if train_data[column].dtype == 'float64':
        float_column.append(column)
print(f'\n Object Column:{object_column}',f'\n Integer Column:{int_column}', f'\n Float Column:{float_column}')
```

```python
int_new_columns = []
for column in int_column:
    if len(train_data[column].unique()) < 15:
        int_new_columns.append(column)
    #print("{}:{}".format(column,train_data[column].unique()))
print(int_new_columns)
```

```python
missing = train_data.isnull().sum().sort_values(ascending=False)*100/len(train_data['Id'])
missing.head(25)
```

```python
train_data.fillna(value=0, axis=1, inplace=True)
train_data.info()
```

```python
train_data.isnull().sum().sort_values(ascending=True).head(25)
```

```python
train_data_after_dummies = pd.get_dummies(train_data, dummy_na=True, columns=object_column)
train_data_after_dummies = pd.get_dummies(train_data_after_dummies, dummy_na=True, columns=int_new_columns)
```

```python
train_data_after_dummies.describe().columns
```

```python
train_data_after_dummies.isnull().sum().sort_values(ascending=False).head(25)
```

```python
train_data_after_dummies.to_csv('processData/train.csv', index=False)
```
